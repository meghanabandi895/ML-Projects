```python
x = df.drop('Outcome',axis=1)
y = df['Outcome']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state=42)
print("x_train ",x_train.shape)
print("x_test ",x_test.shape)
print("y_train ",y_train.shape)
print("y_test ",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='gini',max_depth=3,min_samples_split=10,min_samples_leaf=5,random_state=60)
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score 
print("Model accuracy score with criterion gini index: {0:0.4f}" .format(accuracy_score(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix=confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
from sklearn.metrics import precision_score
pre_score=precision_score(y_test,y_pred,average='macro')
pre_score
```

```python
from sklearn.metrics import recall_score
re_score=recall_score(y_test,y_pred,average='macro')
re_score
```

```python
from sklearn.metrics import f1_score
f1_score=f1_score(y_test,y_pred,average='macro')
f1_score
```

```python
from sklearn.metrics import classification_report
class_report=classification_report(y_test,y_pred)
class_report
formatted_report = "\n".join([line.strip() for line in class_report.split("\n") if line.strip()])
print(formatted_report)
```
